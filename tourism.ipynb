```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Replace 'tourism.csv' with your actual file path
df = pd.read_csv('tourism_dataset.csv')

# Check first few rows
print(df.head())
```

```python
plt.figure(figsize=(12,6))  # Sets the size of the figure
plt.plot(df['Visitors'], marker='o', linestyle='-', color='blue')
plt.title("Visitors Across Locations")
plt.xlabel("Index (Rows in Dataset)")
plt.ylabel("Number of Visitors")
plt.grid(True)
plt.show()
```

```python
countries = df['Country'].unique()  # List of all countries

plt.figure(figsize=(12,6))

for country in countries:
    country_data = df[df['Country'] == country]
    plt.plot(country_data['Visitors'], marker='o', linestyle='-', label=country)

plt.title("Visitors by Country")
plt.xlabel("Index (Rows per Country)")
plt.ylabel("Number of Visitors")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df['Visitors'], df['Revenue'], color='green')
plt.title("Visitors vs Revenue")
plt.xlabel("Visitors")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```
