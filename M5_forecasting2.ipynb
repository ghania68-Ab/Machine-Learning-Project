```python
import pandas as pd

calendar = pd.read_csv("calendar.csv")
sales = pd.read_csv("sales_train_evaluation.csv")
prices = pd.read_csv("sell_prices.csv")

print("Calendar:", calendar.shape)
print("Sales:", sales.shape)
print("Prices:", prices.shape)
```

```python
sales_small = sales.head(20)
print(sales_small.shape)
```

```python
sales_long = sales_small.melt(
    id_vars=['id','item_id','dept_id','cat_id','store_id','state_id'],
    var_name='d',
    value_name='sales'
)
print("Long shape:", sales_long.shape)
sales_long.head()
```

```python
sales_long = sales_long.merge(
    calendar[['d','date','wm_yr_wk','weekday','event_name_1','event_type_1','snap_CA','snap_TX','snap_WI']],
    on='d',
    how='left'
)

print("After calendar merge:", sales_long.shape)
sales_long.head()
```

```python
sales_long = sales_long.merge(
    prices,
    on=['store_id', 'item_id', 'wm_yr_wk'],
    how='left'
)

print("After merging prices:", sales_long.shape)
sales_long.head()
```

```python
sales_long.to_csv("m5_lgta_ready.csv", index=False)
print("Saved final dataset ")
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# ---- LOAD THE CLEAN FILE YOU SAVED ----
m5 = pd.read_csv("m5_lgta_ready.csv")

print("Loaded M5 dataset shape:", m5.shape)

# ---- STEP 1: CREATE A CLEAN TIME SERIES ----
# We will use 'sales' column as the main time series
time_series = m5['sales'].values.astype(float)

print("Raw time series length:", len(time_series))

# ---- STEP 2: SCALE THE TIME SERIES ----
scaler = MinMaxScaler()
time_series_scaled = scaler.fit_transform(time_series.reshape(-1,1)).flatten()

print("Scaled time series length:", len(time_series_scaled))

# ---- STEP 3: CREATE SEQUENCES ----
def create_sequences(data, seq_len=50):
    sequences = []
    for i in range(len(data) - seq_len):
        seq = data[i:i + seq_len]
        sequences.append(seq)
    return np.array(sequences)

X_M5 = create_sequences(time_series_scaled, seq_len=50)

print("Shape of M5 sequences:", X_M5.shape)

# ---- STEP 4: SAVE AS .NPY ----
np.save("M5_forecasting2_preprocessed.npy", X_M5)

print("Saved as M5_forecasting2_preprocessed.npy ")
```

we preprocessed the M5 Forecasting dataset by extracting the 'sales' column, scaling its values between 0 and 1 with MinMaxScaler, and creating overlapping sequences of length 50. The sequences were saved as a .npy file to be used as input for the L-GTA model for generative augmentation experiments.
